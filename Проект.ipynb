# Проект Телекоммуникации

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах. 

**Описание услуг**
    
Оператор предоставляет два основных типа услуг:

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.

2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line, «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).

Также доступны такие услуги:

* Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);

* Выделенная линия технической поддержки (TechSupport);

* Облачное хранилище файлов для резервного копирования данных (OnlineBackup);

* Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

**Описание данных**

Данные состоят из файлов, полученных из разных источников:

* contract_new.csv — информация о договоре;

* personal_new.csv — персональные данные клиента;

* internet_new.csv — информация об интернет-услугах;

* phone_new.csv — информация об услугах телефонии.

Во всех файлах столбец customerID содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

### Замена типа данных

При замене значений `TotalCharges` на float возникает ошибка, поэтому проверим значения.

```python
# выведим значения TotalCharges, которые не удалось преобразовать в числовой тип данных
convertible = pd.to_numeric(contract['TotalCharges'], errors='coerce').notnull()
false_rows = contract[~convertible]
display(false_rows)
```

Как видим, это клиенты у которых контракт начался на дату актуальности данных - 1 февраля 2020 года, следовательно, данные клиенты еще не оплачивали услуги. Заменяем значения на 0 и меняем тип данных. 

```python
contract.loc[contract['TotalCharges'] == " ", 'TotalCharges'] = 0
```

## Моделирование

Данные смешанные (категориальные и числовые), не сбалансированны, поэтому будем применять StandardScaler, OneHotEncoder в Pipeline с помощью ColumnTransformer.

```python
# Объединение предобработчиков признаков
preprocessor = ColumnTransformer(
transformers=[
('numeric', StandardScaler(), numeric_features),
('categorical', OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=-1), categorical_features)
])
```

### RandomForestClassifier

```python
# Построение pipeline
pipeline_RFC = Pipeline(steps=[('preprocessor', preprocessor),
('classifier', RandomForestClassifier(random_state=RANDOM_STATE))])
```

<div class="alert alert-success">
<h2> Комментарий тимлида (код) <a class="tocSkip"> </h2>

<b>Все отлично!👍:</b> Здорово, что используешь сочетание `ColumnTransformer` и `Pipeline` для соединения разных инструментов в единый объект - так твоё решение будет более гибким и масштабируемым. 

```python
param_grid = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [None] + [i for i in range(2, 7)],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4],
}
```

Кроме того, можно попробовать значение `None` - в итоге оптимальная последовательность может выглядеть как `[None] + [i for i in range(2, 7)]`.

### CatBoostClassifier

```python
model = CatBoostClassifier(random_state=RANDOM_STATE)

# Создание pipeline для модели
pipeline_CBC = Pipeline(steps=[('preprocessor', preprocessor), ('model', model)])

# Задание гиперпараметров для подбора
param_grid = [{
    'model__iterations': [100, 150, 200],
    'model__learning_rate': [0.1, 0.01],
    'model__depth': [4, 6, 8]
}]

# Подбор гиперпараметров с использованием кросс-валидации и выбор лучшей модели
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
grid_search_CBC = GridSearchCV(pipeline_CBC, 
                           param_grid, 
                           scoring='roc_auc', 
                           cv=cv,
                           n_jobs=-1)
grid_search_CBC.fit(X_train, y_train)
```
