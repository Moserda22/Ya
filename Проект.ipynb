## 

  (  ),  ,    StandardScaler, OneHotEncoder  Pipeline   ColumnTransformer.

```python
#   
preprocessor = ColumnTransformer(
transformers=[
('numeric', StandardScaler(), numeric_features),
('categorical', OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=-1), categorical_features)
])
```

### RandomForestClassifier

```python
#  pipeline
pipeline_RFC = Pipeline(steps=[('preprocessor', preprocessor),
('classifier', RandomForestClassifier(random_state=RANDOM_STATE))])
```

<div class="alert alert-success">
<h2>   () <a class="tocSkip"> </h2>

<b> !:</b> ,    `ColumnTransformer`  `Pipeline`        -        . 

```python
param_grid = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [None] + [i for i in range(2, 7)],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4],
}
```

<div class="alert alert-warning">
    <h2>   <a class="tocSkip"> </h2>

```python
#    -    
cv = StratifiedKFold(n_splits=5)
grid_search_RFC = GridSearchCV(estimator=pipeline_RFC,
                           param_grid=param_grid, 
                           cv=cv, 
                           scoring='roc_auc',
                           n_jobs=-1,
                           )
grid_search_RFC.fit(X_train, y_train)
```

```python
best_params_RFC=grid_search_RFC.best_params_
best_score_RFC=grid_search_RFC.best_score_
```

```python
print(" :", best_params_RFC)
print("Cross-validation AUC-ROC:", best_score_RFC)
```

### CatBoostClassifier

```python
model = CatBoostClassifier(random_state=RANDOM_STATE)

#  pipeline  
pipeline_CBC = Pipeline(steps=[('preprocessor', preprocessor), ('model', model)])

#    
param_grid = [{
    'model__iterations': [100, 150, 200],
    'model__learning_rate': [0.1, 0.01],
    'model__depth': [4, 6, 8]
}]

#     -    
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
grid_search_CBC = GridSearchCV(pipeline_CBC, 
                           param_grid, 
                           scoring='roc_auc', 
                           cv=cv,
                           n_jobs=-1)
grid_search_CBC.fit(X_train, y_train)
```

```python
best_params_CBC=grid_search_CBC.best_params_
best_score_CBC=grid_search_CBC.best_score_
```

```python
#     auc-roc   
print("Best hyperparameters: ", best_params_CBC)
print("AUC-ROC on training set: ", best_score_CBC)
```
